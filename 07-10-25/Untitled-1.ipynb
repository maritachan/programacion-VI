## MARCO REFERENCIAL DE PROYECTO PRACTICO FINAL

EL OBJETIVO ES LOGRAR QUE LOS INTEGRANTES INTEGREN LOS CONOCIMIENTOS ADQUIRIDOS MEDIANTE EL DESARROLLO DE UNA APLICACION FUNCIONAL, ESTETICA Y CON LOGICA DE DATOS COHERENTE.

## 2- Alcance y tipologia de los proyectos

El proyecto debe ser una aplicacion de escritorio o web desarrollada en Flet, con un minimo de tres (3), pantallas (vistas), y almacenamiento de datos (SQLite)

### Tipos de proyectos sugeridos

- Sistema de gestion (Inventarios, biblioteca, gastos, tareas, notas)
- Aplicacion tematica (Salud, Educacion, cultura, musica, deportes)
- Mini red social o blog
- App con elementos visuales interactivos (graficos, animaciones, formularios dinamicos)
CADA ALUMNO DEBE DE PRESENTAR UNA IDEA ORIGINAL VALIDADA POR EL DOCENTE MEDIANTE UNA FICHA DE PROYECTO

## 3- Fases del proyecto (evaluacion secuencial)

3.1 - Ideacion y estructura. Carga de 10% (trabajo practico) 
Entrega de ficha con el nombre del proyecto, objetivos, publico objetivo, prototipo (en canva, wireframe en papel o digital), descripcion de vistas y fuciones. EN PDF EL PROTOTIPO 

3.2 - Implementacion de base:(Segunda Parcial) 15%. App funcional con navegacion entre pantallas, formularios, botones y eventos implementados.
Se va a evaluar codigo limpio y la estructura modal.
3.3 - Persistencia de datos: (FINAL) 60%.  Integracion de base de datos o almacenamiento persistente. Validacion de entrada, interfas mejorada y entrega final ejecutable. 

## 4- Reglas de autenticidad y originalidad

1. Prohibido clonar proyectos de internet. El codigo sera revisado con herramientas de similitud (Ej. GitHub, Copilot, Compare, diffecheker y la IA en general)

2. Cada alumno debe de entregar un documento de respaldo (PDF):
    - Portada institucional
    - Descripcion del proyecto
    - Capturas de pantallas propias
    - Explicacion breve del codigo
    - Conclusiones personales 

3. Defensa individual obligatoria: explicacion del funcionamiento, desafios y mejoras posibles.

4. El proyecto debe de ejecutarse correctamente en una PC estandar o entorno offline sin depender de servicios externos
